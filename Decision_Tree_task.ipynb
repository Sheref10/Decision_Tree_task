```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
import pandas_profiling as pp
import matplotlib
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from dtreeviz.trees import dtreeviz
```

#Load the dataset

```python
iris=datasets.load_iris()
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)
target=iris.target
target
```

#Data exploration


```python
print(iris_data.info)
print(iris_data.describe())
```

#Data Cleaning

```python
print(iris_data.duplicated().sum())
print(iris_data.duplicated().sum())
print(iris_data.isnull().sum())
iris_data.corr().style.background_gradient(cmap='green')
iris_data.boxplot()
```

# Modeling the decision tree algorithm

```python
Decision_Tree=DecisionTreeClassifier()
Decision_Tree.fit(iris_data,target)
print('Decision Tree Classifer Created')
```

#Decision Tree visualization

```python
fig = plt.figure(figsize=(25,20))
viz = tree.plot_tree(Decision_Tree,
                   feature_names=iris.feature_names,
                   class_names=iris.target_names,
                   filled=True)
viz
```

```python
viz= dtreeviz(Decision_Tree, iris_data, target,
                target_name="target",
                feature_names=iris.feature_names,
                class_names=list(iris.target_names),orientation='LR')

viz
```

```python
X = iris.data[np.random.randint(0, len(iris.data)),:]

viz = dtreeviz(Decision_Tree,
               iris_data,
               target,
               target_name='target',  # this name will be displayed at the leaf node
               feature_names=iris.feature_names,
               X=X, 
               show_node_labels = True
              )
viz
```
